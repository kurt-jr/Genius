```python
import numpy as np
import pandas as pd
import requests
import json
import os
import dotenv
import sys
import lyricsgenius # a wrapper around requests that works specifically with the Genius API
sys.tracebacklimit = 0 # turn off the error tracebacks
```

Genius API Documentation: https://docs.genius.com/#/getting-started-h1

```python
#Accessing the Genius API with the clientID, secretClient, and accessToken from an .env file
dotenv.load_dotenv()
clientID = os.getenv('clientID')
secretClient = os.getenv('secretClient')
accessToken = os.getenv('accessToken')
```

```python
r =requests.get('https://httpbin.org/user-agent')
useragent = json.loads(r.text)['user-agent']
headers = {'User-agent': useragent, 'From': "Kurt Icarus"}
```

```python
params = {'access_token':accessToken,
         'q': "Don Toliver"}
root = 'https://api.genius.com'
endpoint = '/search'
r =requests.get(root+endpoint, params = params, headers = headers)
r
```

```python
DonT_hits = pd.json_normalize(json.loads(r.text), record_path=['response', 'hits'])
DonT_hits
```

```python
params2 = {'access_token':accessToken,
         'q': "Childish Gambino", 'per_page': 25}
c =requests.get(root+endpoint, params = params2, headers = headers)
CG_hits = pd.json_normalize(json.loads(c.text), record_path=['response', 'hits'])
```
